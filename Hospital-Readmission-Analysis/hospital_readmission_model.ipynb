```python
# How many patients were readmitted vs not
query = """
SELECT readmitted, COUNT(*) AS total
FROM df
GROUP BY readmitted
"""
ps.sqldf(query, locals())
```

```python
# Most common number of medications given to patients
query = """
SELECT num_medications, COUNT(*) AS patient_count
FROM df
GROUP BY num_medications
ORDER BY patient_count DESC
LIMIT 5
"""
ps.sqldf(query, locals())
```

```python
import pandas as pd

# Reload dataset (in case)
df = pd.read_csv("train[1].csv")   # make sure filename matches your uploaded file

# 1. Check shape
print("Rows, Columns:", df.shape)

# 2. See missing values
print("\nMissing values per column:")
print(df.isnull().sum().sort_values(ascending=False).head(10))

# 3. Quick summary of numeric columns
print("\nDataset description:")
print(df.describe().T.head(10))
```

```python
# 1. Drop columns that are not useful
# (we donâ€™t have ID or Name in this dataset, but letâ€™s double-check)
print("Columns:", df.columns[:20].tolist())  # show first 20 column names

# 2. Convert booleans to integers (True/False â†’ 1/0)
df_clean = df.copy()
df_clean = df_clean.astype(int, errors='ignore')

# 3. Check final dataset shape and types
print("\nShape after cleaning:", df_clean.shape)
print("\nColumn types:")
print(df_clean.dtypes.head(10))

```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report

# Features (all columns except readmitted)
X = df_clean.drop("readmitted", axis=1)

# Target (the column we want to predict)
y = df_clean["readmitted"]

# Split into train (70%) and test (30%)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Train logistic regression model
model = LogisticRegression(max_iter=2000)
model.fit(X_train, y_train)

# Predictions
y_pred = model.predict(X_test)

# Evaluate model
print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```

```python
# Add predictions to test dataset
df_results = X_test.copy()
df_results["actual_readmitted"] = y_test
df_results["predicted_readmitted"] = y_pred

# Save for dashboard
df_results.to_csv("hospital_predictions.csv", index=False)

print("âœ… Predictions saved as hospital_predictions.csv")
df_results.head()

```
